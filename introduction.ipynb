### MACHINE LEARNING

This approach involves setting up models or algorithms that allows the machine to learn from historical data. We train our historical dataset on the model and we fit the model on the dataset as best as possible then use this newly created model to predict the outcome of future events that it has never seen based on the patterns that it has learnt from the historical dataset


1. #### SUPERVISED LEARNING 

This is the type of machine learning algorithms that trains a model on a dataset with the purpose of predicting a  response variable. These response variables usually exist in the form of dependent variables.

#### Classes of Supervised learning algorithms

* #### Regression
This type of supervised learning algorithm learns from the dataset to predict a continuous outcome/response variable. Example of these algorithms include linear regression, decision tree regression, xgboost regression etc


* #### Classification
This type of supervised learning algorithm learns from the dataset to predict a categorical outcome/response variable. Example of these algorithms include logistic regression, decision tree classifier, xgboost classifier etc


2. #### Unsupervised Learning
In this type of machine learning, the model isi trained to learn from dataset without a target variable. The algorithm is structured in a way that the model learns pattern from the dataset without seeing a dependent variable.
Example includes K means clustering, Heirachical clustering etc


3. #### Reinforcement Learning
This type of machine learning the agent is allowed to learn from a defined system under a penalty and reward scope.
Example Q learning, Deep Q learning, Proximal Policy Optimization etc




### Lifecycle of a Machine Learning Project

1. ##### Source for the data
    * from sql database
    * through the web (kaggle, uci machine learning repository, worldbank)
    * through web scraping (beautifulsoup, requests, selenium, scrapy)
    * through APIs (tweepy, yfinance, coinmarketapp, nba)
    * load it locally
2. ##### Exploratory Data Analysis
    * Data Wrangling (missing values, incorrect data types, outlier detection, transformation etc)
    * Data Visualization 
3. #### Train-Test-Validation Split
    This is the level where we split our dataset into 3 categories; The training set, the test set and the validation set.
    * Training set:
        Usually takes above 70% of the entire dataset. This are the records that we use to train the model
    * Validation set:
        This is usually taken from the training set and used for validating the model performance before exposing it to the test set.
    * Test set:
        This is usually around 20% of the data and used for testing the model performance before exposing it to on premise scenario.
    
4. #### Model Training
    This is the stage where we train the model on the training set using the right algorithm

5. #### Model Evaluation
    This is the level where we inspect the performance of the model. To see if the model is a right fit for predicting on premise scenario.

6. # Hyperparameter Optimization
    It is the stage where we make our model better by tuning the algorithm's hyperparameter to either minimize the loss or improve the accuracy. Examples of libraries include Gridsearch, RandomizedSearch, Optuna etc

7. # Deploy the Model.
    This is the final stage where we allow end users to interact with our model. Models can be deployed in form of :
        1. A demo website or an MVP. Streamlit and Gradio
        2. API using FASTAPI, Flask, Django RestAPI etc
